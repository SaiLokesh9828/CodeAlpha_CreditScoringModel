```python
import pandas as pd
import numpy as np
credit_df = pd.read_csv("train.csv")
```

```python
credit_df.columns
```

```python
credit_df.head()
```

```python
#drop unnecessary features
df = credit_df.drop(['ID', 'Customer_ID', 'Month', 'Name','Type_of_Loan','Credit_History_Age','SSN'],axis='columns')
df.head()
```

```python
#Check for object data type columns
df.select_dtypes(include=['object']).columns
```

```python
#let us convert non categorical object columns into numerical 
df['Age']=pd.to_numeric(df['Age'],errors='coerce')
df['Annual_Income']=pd.to_numeric(df['Annual_Income'],errors='coerce')
df['Num_of_Loan']=pd.to_numeric(df['Num_of_Loan'],errors='coerce')
df['Num_of_Delayed_Payment']=pd.to_numeric(df['Num_of_Delayed_Payment'],errors='coerce')
df['Changed_Credit_Limit']=pd.to_numeric(df['Changed_Credit_Limit'],errors='coerce')
df['Outstanding_Debt']=pd.to_numeric(df['Outstanding_Debt'],errors='coerce')
df['Amount_invested_monthly']=pd.to_numeric(df['Amount_invested_monthly'],errors='coerce')
df['Monthly_Balance']=pd.to_numeric(df['Monthly_Balance'],errors='coerce')
```

```python
df.select_dtypes(include=['object']).columns
```

```python
#convert invalid columns of age column into nan
df.loc[df['Age']<=0,'Age'] = np.nan
df[df['Age']<=0]
```

```python
#let us find missing values
df.isnull().sum()
```

```python
# fill the missing values with median
df['Age']=df['Age'].fillna(df['Age'].median())
df['Annual_Income']=df['Annual_Income'].fillna(df['Annual_Income'].median())
df['Monthly_Inhand_Salary']=df['Monthly_Inhand_Salary'].fillna(df['Monthly_Inhand_Salary'].median())
df['Num_of_Loan']=df['Num_of_Loan'].fillna(df['Num_of_Loan'].median())
df['Num_of_Delayed_Payment']=df['Num_of_Delayed_Payment'].fillna(df['Num_of_Delayed_Payment'].median())
df['Changed_Credit_Limit']=df['Changed_Credit_Limit'].fillna(df['Changed_Credit_Limit'].median())
df['Num_Credit_Inquiries']=df['Num_Credit_Inquiries'].fillna(df['Num_Credit_Inquiries'].median())
df['Outstanding_Debt']=df['Outstanding_Debt'].fillna(df['Outstanding_Debt'].median())
df['Amount_invested_monthly']=df['Amount_invested_monthly'].fillna(df['Amount_invested_monthly'].median())
df['Monthly_Balance']=df['Monthly_Balance'].fillna(df['Monthly_Balance'].median())
```

```python
df.isnull().sum()
```

```python
#Encoding categorical columns
```

```python
from sklearn.preprocessing import LabelEncoder
le_Occupation = LabelEncoder()
le_Credit_Mix = LabelEncoder()
le_Payment_of_Min_Amount = LabelEncoder()
le_Payment_Behaviour = LabelEncoder()
le_Credit_Score = LabelEncoder()
```

```python
df['Occupation_n']=le_Occupation.fit_transform(df['Occupation'])
df['Credit_Mix_n']=le_Credit_Mix.fit_transform(df['Credit_Mix'])
df['Payment_of_Min_Amount_n']=le_Payment_of_Min_Amount.fit_transform(df['Payment_of_Min_Amount'])
df['Payment_Behaviour_n']=le_Payment_Behaviour.fit_transform(df['Payment_Behaviour'])
df['Credit_Score_n']=le_Credit_Score.fit_transform(df['Credit_Score'])
```

```python
df.head()
```

```python
#drop object categorical columns
df1 = df.drop(['Occupation','Credit_Mix','Payment_of_Min_Amount','Payment_Behaviour','Credit_Score'],axis='columns')
df1.head()
```

```python
#split data into train and test datasets
```

```python
from sklearn.model_selection import train_test_split
X = df1.drop(['Credit_Score_n'],axis='columns')
y = df1.Credit_Score_n
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
#Use Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
lr_model.score(X_test,y_test)
```

```python
y_pred_lr = lr_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred_lr))
```

```python
#Use Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(X_train,y_train)
```

```python
dt_model.score(X_test,y_test)
```

```python
y_pred_dt = dt_model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_dt))
```

```python
#Use Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
```

```python
rf_model.fit(X_train,y_train)
```

```python
rf_model.score(X_test,y_test)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_rf))
```
